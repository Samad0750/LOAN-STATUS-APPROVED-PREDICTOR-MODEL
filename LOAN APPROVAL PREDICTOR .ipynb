```python
df['ApplicantIncome_log'] = np.log1p(df['ApplicantIncome'])
df['LoanAmount_log'] = np.log1p(df['LoanAmount'])
df['Loan_Amount_Term_log'] = np.log1p(df['Loan_Amount_Term'])
df['Toatl Income_log'] = np.log1p(df['Toatl Income'])
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
df['Married']=df['Married'].map({'No':0,'Yes':1})
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0})
df['Property_Area']=df['Property_Area'].map({'Urban':2,'Rural':0,'Semiurban':1})
df['Self_Employed']=df['Self_Employed'].map({'No':0,'Yes':1})
df['Loan_Status']=df['Loan_Status'].map({'Y':1,'N':0})
```

```python
df['Credit_History']=df['Credit_History'].astype(int)
```

```python
df['Dependents']=df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3,1:1})
```

```python

columns = [col for col in df.columns if col != 'Loan_Status']

# Add 'Loan_Status' to the end of the list of columns
columns.append('Loan_Status')

# Reorder the DataFrame columns
df = df[columns]

```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=df['Loan_Status']
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
# Define the classifiers and their parameter grids for randomized search
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Define the classifiers with their parameter grids
classifiers = {
    'Logistic Regression': (LogisticRegression(), {
        'C': [0.1, 1, 10],
        'solver': ['lbfgs', 'liblinear'],
        'max_iter': [100, 200],
    }),
    'Decision Tree': (DecisionTreeClassifier(), {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 10, 20],
    }),
    'Random Forest': (RandomForestClassifier(), {
        'n_estimators': [10, 50, 100],
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 10, 20],
    }),
    'K-Nearest Neighbors': (KNeighborsClassifier(), {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
    }),
    'Support Vector Machine': (SVC(), {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'poly', 'rbf'],
    }),
    'Neural Network': (MLPClassifier(), {
        'hidden_layer_sizes': [(100,), (50, 50)],
        'activation': ['relu', 'tanh'],
        'max_iter': [200, 300],
    })
}

# Perform randomized search on each classifier
results = {}
for clf_name, (clf, param_grid) in classifiers.items():
    random_search = RandomizedSearchCV(clf, param_grid, n_iter=10, cv=3, random_state=42, n_jobs=-1)
    random_search.fit(X_train, y_train)
    
    # Get the best estimator and make predictions
    best_clf = random_search.best_estimator_
    y_pred = best_clf.predict(X_test)
    
    # Compute the confusion matrix and accuracy
    conf_matrix = confusion_matrix(y_test, y_pred)
    acc_score = accuracy_score(y_test, y_pred)
    
    # Store the results
    results[clf_name] = {
        'best_estimator': best_clf,
        'confusion_matrix': conf_matrix,
        'accuracy': acc_score,
    }

# Display the results with confusion matrices and accuracy scores
for clf_name, result in results.items():
    print(f"\nClassifier: {clf_name}")
    print(f"Best Parameters: {result['best_estimator'].get_params()}")
    print(f"Accuracy Score: {result['accuracy']:.2f}")
    
    ConfusionMatrixDisplay.from_predictions(y_test, result['best_estimator'].predict(X_test),cmap='Blues', values_format='d')
    plt.title(f"Confusion Matrix - {clf_name}")
    plt.show()
```
